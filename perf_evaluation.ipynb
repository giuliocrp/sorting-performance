```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read the data
quick_data = pd.read_csv('quicksort_times.csv')
bubble_data = pd.read_csv('bubblesort_times.csv')
```

```python
# Plotting Bubble Sort data
plt.figure(figsize=(10,6))
plt.plot(bubble_data['n'], bubble_data['time_ms'], label='Bubble Sort', marker='o')

# Plot O(n^2) reference
n_values = bubble_data['n']
onsquare = n_values ** 2
onsquare = onsquare / onsquare.max() * bubble_data['time_ms'].max()  # Normalize for comparison
plt.plot(n_values, onsquare, '--', label='O(n^2) reference')

plt.xlabel('Array Size n')
plt.ylabel('Time (ms)')
plt.title('Bubble Sort Performance')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting QuickSort data
plt.figure(figsize=(10,6))
plt.plot(quick_data['n'], quick_data['time_ms'], label='QuickSort', marker='o')

# Plot O(n log n) reference
n_values = quick_data['n']
onlogn = n_values * np.log2(n_values)
onlogn = onlogn / onlogn.max() * quick_data['time_ms'].max()  # Normalize for comparison
plt.plot(n_values, onlogn, '--', label='O(n log n) reference')

plt.xlabel('Array Size n')
plt.ylabel('Time (ms)')
plt.title('QuickSort Performance')
plt.legend()
plt.grid(True)
plt.show()
```
